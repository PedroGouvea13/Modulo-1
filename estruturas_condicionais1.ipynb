```python
# IF/Else
num1 = 10
num2 = 10
if num1 > num2:
  print("Num1 é maior")
elif:
  print("Num2 é maior")
```

```python
# IF/ELIF/ELSE

num1 = 100
num2 = 14
if num1 > num2:
  print("num1 e o maior. ")
elif num1 < num2:
  print("num2 e o maior.")
else:
  print("Os dois sao iguais.")
```

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maio Numero.')
else:
  print("O num1 nao é o maior Numero.")
```
